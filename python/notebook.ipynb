```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../data/workloads_with_recommendations_V2.csv")

print(f"Shape : {data.shape[0]} lignes x {data.shape[1]} colonnes\n")

print("=== Aperçu ===")
display(data.head())

print("\n=== Types des colonnes ===")
print(data.dtypes)

print("\n=== Valeurs manquantes ===")
missing = data.isnull().sum()
missing = missing[missing > 0]
if missing.empty:
    print("Aucune valeur manquante")
else:
    print(missing)

print("\n=== Valeurs uniques par colonne ===")
print(data.nunique())

print("\n=== Statistiques descriptives (colonnes numériques) ===")
display(data.describe())

print("\n=== Statistiques descriptives (colonnes catégorielles) ===")
display(data.describe(include="object"))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution de l'énergie originale vs estimée
axes[0, 0].hist(data["orig_energy_kWh"], bins=50, alpha=0.6, label="Originale")
axes[0, 0].hist(data["est_energy_kWh"], bins=50, alpha=0.6, label="Estimée")
axes[0, 0].set_yscale("log")
axes[0, 0].set_xlabel("Énergie (kWh)")
axes[0, 0].set_ylabel("Nombre de jobs (log)")
axes[0, 0].set_title("Distribution de l'énergie consommée")
axes[0, 0].legend()

# 2. Distribution de la durée originale vs estimée
axes[0, 1].hist(data["orig_duration_h"], bins=50, alpha=0.6, label="Originale")
axes[0, 1].hist(data["est_duration_h"], bins=50, alpha=0.6, label="Estimée")
axes[0, 1].set_yscale("log")
axes[0, 1].set_xlabel("Durée (h)")
axes[0, 1].set_ylabel("Nombre de jobs (log)")
axes[0, 1].set_title("Distribution de la durée des jobs")
axes[0, 1].legend()

# 3. Puissance moyenne : originale vs estimée (scatter)
axes[1, 0].scatter(data["orig_avg_power_W"], data["est_avg_power_W"], alpha=0.15, s=5)
lim = max(data["orig_avg_power_W"].max(), data["est_avg_power_W"].max())
axes[1, 0].plot([0, lim], [0, lim], "r--", label="x = y")
axes[1, 0].set_xlabel("Puissance orig. (W)")
axes[1, 0].set_ylabel("Puissance est. (W)")
axes[1, 0].set_title("Puissance moyenne : originale vs estimée")
axes[1, 0].legend()

# 4. Nombre de jobs par cluster original
cluster_counts = data["orig_cluster"].value_counts().sort_index()
axes[1, 1].bar(cluster_counts.index, cluster_counts.values)
axes[1, 1].set_xlabel("Cluster")
axes[1, 1].set_ylabel("Nombre de jobs")
axes[1, 1].set_title("Répartition des jobs par cluster original")
axes[1, 1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 5. Matrice de flux : cluster original → cluster recommandé
migration = data.groupby(["orig_cluster", "recommended_cluster"]).size().unstack(fill_value=0)
im = axes[0].imshow(migration.values, cmap="YlOrRd", aspect="auto")
axes[0].set_xticks(range(len(migration.columns)))
axes[0].set_xticklabels(migration.columns, rotation=45, ha="right")
axes[0].set_yticks(range(len(migration.index)))
axes[0].set_yticklabels(migration.index)
axes[0].set_xlabel("Cluster recommandé")
axes[0].set_ylabel("Cluster original")
axes[0].set_title("Matrice de migration entre clusters")
plt.colorbar(im, ax=axes[0], label="Nombre de jobs")

# 6. Gain d'énergie par cluster original
data["energy_savings_kWh"] = data["orig_energy_kWh"] - data["est_energy_kWh"]
savings_by_cluster = data.groupby("recommended_cluster")["energy_savings_kWh"].mean().sort_index()
colors = ["green" if v > 0 else "red" for v in savings_by_cluster.values]
axes[1].bar(savings_by_cluster.index, savings_by_cluster.values, color=colors)
axes[1].axhline(0, color="black", linewidth=0.8)
axes[1].set_xlabel("Cluster recommandé")
axes[1].set_ylabel("Économie moyenne (kWh)")
axes[1].set_title("Économie d'énergie moyenne par cluster")
axes[1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ===== Tableau récapitulatif des 8 scénarios =====
# On regroupe les bilans pour comparer sans/avec backfill côte à côte.

recap8 = pd.DataFrame({
    "S1 (100% reco)":                bilan,
    "S1 + backfill":                 bilan_backfill,
    "S2 (50% acceptance)":           bilan2,
    "S2 + backfill":                 bilan2_backfill,
    "S3 (80% acceptance)":           bilan3,
    "S3 + backfill":                 bilan3_backfill,
    "S4 (max_wait=2h)":              bilan4,
    "S4 + backfill":                 bilan4_backfill,
    "S5 (60% rate_reco)":            bilan5,
    "S5 + backfill":                 bilan5_backfill,
}).T

cols = [
    "nb_jobs",
    "gain_energie_%",
    "attente_moyenne_h",
    "attente_max_h",
    "nb_migrations",
    "taux_adhesion_effectif_%",
    "nb_timeouts_fallback",
]
display(recap8[cols].round(2))
```

```python
# ===== Apport du backfill par scénario =====
# Pour chaque scénario, on calcule la différence entre la version avec et sans backfill
# sur les 3 métriques qui comptent : gain énergie, attente moyenne, migrations.

paires = [
    ("S1 (100% reco)",      bilan,  bilan_backfill),
    ("S2 (50% acceptance)", bilan2, bilan2_backfill),
    ("S3 (80% acceptance)", bilan3, bilan3_backfill),
    ("S4 (max_wait=2h)",    bilan4, bilan4_backfill),
    ("S5 (60% rate_reco)",  bilan5, bilan5_backfill),
]

delta = pd.DataFrame([
    {
        "Scénario": nom,
        "Δ gain énergie (pts)": round(bf["gain_energie_%"] - sans["gain_energie_%"], 2),
        "Δ attente moy (h)":    round(bf["attente_moyenne_h"] - sans["attente_moyenne_h"], 3),
        "Δ migrations":         int(bf["nb_migrations"] - sans["nb_migrations"]),
    }
    for nom, sans, bf in paires
])
display(delta)
print("\nLecture : un Δ positif sur le gain = le backfill améliore. Un Δ négatif sur l'attente = le backfill réduit l'attente.")
```

```python
# ===== Graphiques comparatifs : sans vs avec backfill =====
# 5 scénarios x 2 versions, regroupés deux par deux pour lisibilité.

import numpy as np

labels = ["S1", "S2", "S3", "S4", "S5"]
gains_sans = [bilan["gain_energie_%"], bilan2["gain_energie_%"], bilan3["gain_energie_%"], bilan4["gain_energie_%"], bilan5["gain_energie_%"]]
gains_bf   = [bilan_backfill["gain_energie_%"], bilan2_backfill["gain_energie_%"], bilan3_backfill["gain_energie_%"], bilan4_backfill["gain_energie_%"], bilan5_backfill["gain_energie_%"]]
atts_sans  = [bilan["attente_moyenne_h"], bilan2["attente_moyenne_h"], bilan3["attente_moyenne_h"], bilan4["attente_moyenne_h"], bilan5["attente_moyenne_h"]]
atts_bf    = [bilan_backfill["attente_moyenne_h"], bilan2_backfill["attente_moyenne_h"], bilan3_backfill["attente_moyenne_h"], bilan4_backfill["attente_moyenne_h"], bilan5_backfill["attente_moyenne_h"]]

x = np.arange(len(labels))
width = 0.35

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].bar(x - width/2, gains_sans, width, label="sans backfill", color="#3498db")
axes[0].bar(x + width/2, gains_bf,   width, label="avec backfill", color="#e67e22")
axes[0].set_xticks(x); axes[0].set_xticklabels(labels)
axes[0].set_ylabel("Gain énergie (%)")
axes[0].set_title("Gain énergétique par scénario")
axes[0].legend()
for i, (a, b) in enumerate(zip(gains_sans, gains_bf)):
    axes[0].text(i - width/2, a + 0.3, f"{a:.1f}", ha="center", fontsize=8)
    axes[0].text(i + width/2, b + 0.3, f"{b:.1f}", ha="center", fontsize=8)

axes[1].bar(x - width/2, atts_sans, width, label="sans backfill", color="#3498db")
axes[1].bar(x + width/2, atts_bf,   width, label="avec backfill", color="#e67e22")
axes[1].set_xticks(x); axes[1].set_xticklabels(labels)
axes[1].set_ylabel("Attente moyenne (h)")
axes[1].set_title("Attente moyenne par scénario")
axes[1].legend()

plt.tight_layout()
plt.show()
```
